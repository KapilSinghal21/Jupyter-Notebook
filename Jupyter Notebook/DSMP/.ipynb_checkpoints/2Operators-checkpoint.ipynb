```python
# Menu Driven Calculator

fnum = int ( input ('Enter Num1'))
snum = int ( input ('Enter NUm2'))
op = input ('Enter Operation')

if op == '+':
    print( fnum+snum )
elif op == '-':
    print( fnum-snum )
elif op == '*':
    print(fnum*snum)
else:
    print(fnum/snum)
    
```

# 3 Loops in Python ( While Loop & for Loop ) 

# Loops = repeated piece of code 
IMP # Need for Loops Display of information Product from container -> database (loops) Ex - Flipkart , Zomato

```python
# While Loop 
Number = int(input('Enter your number'))
i = 1
while i<11 :
    print(Number , '*' , i , '=' , Number * i) # or just Number * i
    i += 1
```

```python
# While loop with Else
x = 1
while x<3:
    print(x)
    x+=1
else:
    print('Limit Crossed')
```

```python
# for loop 
for i in {1,2,3,4,5}:
    print(i)
```

# If we have to write end condition we write while loop , if we have to iterate No end condition we use For loop , We can use if else in for loop in python .

# Nested Loops

```python
for i in range(1,5):  # 1 se start , 5 is not included
    for j in range(1,5):
        print(i,j)
```

# Interview Q

```python
# Sequence Sum (Interview)
n = int(input('enter n'))

result = 0
fact = 1

for i in range(1,n+1):
  fact = fact * i
  result = result + i/fact

print(result)
```

```python
# Pattern 1
'''
*
**
***
****
*****
'''
rows = int(input('enter numbers of rows'))

for i in range (1,rows+1):
    for j in range (1,i+1):
        print('*' , end = ' ')
    print()
```

```python
# Pattern 2 
''' 
1
1 2 1 
1 2 3 2 1
1 2 3 4 3 2 1
''' 

rows = int (input ('Enter no of rows'))

for i in range ( 1 , rows+1):
    for j in range (1 , i+1):
        print(j , end = ' ')
    for k in range (i-1 , 0-1):
        print(k, end = '')
    print()
```

# Loop Control Statements Break Continue Pass

```python
for i in range (1,10):
    if i == 5:
        break 
    print(i) # Indentation (keep in mind) 

# Linear Searching
```

```python
# Used in Database when search something than we can break 
# Ex : Searching a name in thousands of data we use break to not check rest 
```

```python
# Q User will give you range print prime no b/w them 
# Python has Intendation and o/p will change by silly error

lower = int(input('Enter Lower range: '))
upper = int(input('Enter Upper range: '))

for i in range(lower, upper):
    if i > 1:
        for j in range(2, i):
            if i % j == 0:
                break
        else:
            print(i)
```

```python
for i in range(1,10):
    if i == 5:
        continue # Skip Current Iteration ( Ecommerce Website Skip product which are out of stock )
    print(i)
```
